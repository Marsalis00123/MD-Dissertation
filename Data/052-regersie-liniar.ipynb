```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Încarcă fișierul CSV
file_path = "anonim-clean.csv"  # Asigură-te că fișierul este în același director cu notebook-ul

df = pd.read_csv(file_path)

# Verifică primele rânduri și tipurile de date
display(df.info())
display(df.head())

# Calculează numărul de comorbidități (număr de elemente separate prin virgulă în coloana 'app')
df['nr_comorbiditati'] = df['app'].fillna('').apply(lambda x: len(x.split(',')) if x else 0)

# Calculează numărul de investigații imagistice (număr de metode separate prin virgulă în 'metodă imagistică')
df['nr_investigatii'] = df['metodă imagistică'].fillna('').apply(lambda x: len(x.split(',')) if x else 0)

# Calculează coeficientul de corelație Pearson
correlation = df[['nr_comorbiditati', 'nr_investigatii']].corr().iloc[0, 1]
print(f"Coeficientul de corelație Pearson: {correlation:.3f}")

# Definim variabilele pentru regresie
X = df['nr_comorbiditati']  # Variabila independentă (predictor)
y = df['nr_investigatii']   # Variabila dependentă

# Adăugăm o coloană de 1 pentru intercept
X = sm.add_constant(X)

# Construim modelul de regresie liniară
model = sm.OLS(y, X).fit()

# Afișăm rezumatul modelului
print(model.summary())

# Vizualizare: scatterplot + linia de regresie
plt.figure(figsize=(8, 6))
sns.regplot(x=df['nr_comorbiditati'], y=df['nr_investigatii'], ci=None, line_kws={'color': 'red'})
plt.xlabel("Număr Comorbidități")
plt.ylabel("Număr Investigații Imagistice")
plt.title("Regresie Liniară: Comorbidități vs Investigații")
plt.grid()
plt.show()
```
